```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 4, 5, 6])

# Crear el modelo de regresi칩n lineal
model = LinearRegression()

# Ajustar el modelo a los datos
model.fit(x.reshape(-1, 1), y)

# Obtener la pendiente (m) e intercepto (b)
m = model.coef_[0]
b = model.intercept_

# Crear la nube de puntos
plt.scatter(x, y, color='blue', label='Datos')

# Crear la recta de regresi칩n
plt.plot(x, m*x + b, color='red', label='Recta de regresi칩n')

# Etiquetas y leyenda
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Mostrar el gr치fico
plt.show()
```
